**Exercício 1: Filtragem de dados**

Filtre o Dataframe a partir de 5 ações, de modo que essas ações podem ser escolhidas pela sua preferência ou de forma aleatória. Após a filtragem, salve em uma nova variável.

```python
##### SEU CODIGO COMECA AQUI #####
import random

resp = input("Gostaria de Escolher as Ações? (Sim/Não) ")

if resp =="Sim":
    i = 1
    acoes = [input("Digite a primeira ação: ")]
    while i < 5:
        acao = input("Digite a próxima ação: ")
        if acao in (at.columns):
            acoes.append(acao)
            i += 1
        else:
            print("Essa ação não está no banco de dados.")
else:
    filt_aux = at.filter(regex=r'^.{5}$', axis=1)
    acoes = random.sample(list(filt_aux.columns), 5)

at_filtr = at.filter(acoes, axis=1)
```

**Exercício 2: Retorno**

Utilizando o Dataframe filtrado no exercício anterior, calcule o retorno diário dessas 5 ações. Para isso, utilize o método `pct_change`. Além disso, mude o nome das colunas para o nome da ação seguido de uma hashtag, por exemplo, 'BBAS3#'.

salve um uma variável chamada `retorno`.

```python
##### SEU CODIGO COMECA AQUI #####
novos_nomes = []
for elemento in at_filtr.columns:
    novos_nomes.append(f"{elemento}#")

retorno = at_filtr.pct_change().fillna(0).rename(columns={nome_atual: novo_nome 
                                                   for nome_atual, novo_nome 
                                                   in zip(at_filtr.columns, novos_nomes)})
```
